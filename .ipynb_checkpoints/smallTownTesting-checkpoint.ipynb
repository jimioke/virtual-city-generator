```python
def kde1(x, y, ax):
    from scipy.stats import gaussian_kde
    
    # Calculate the point density
    xy = np.vstack([x,y])
    kernel = gaussian_kde(xy, bw_method='silverman')

    xmin = x.min()
    xmax = x.max()
    ymin = y.min()
    ymax = y.max()
    
    X, Y = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
    positions = np.vstack([X.ravel(), Y.ravel()])

    Z = np.reshape(kernel(positions).T, X.shape)

    ax.imshow(np.rot90(Z), cmap=plt.cm.viridis,
              extent=[xmin, xmax, ymin, ymax])
    
    ax.scatter(x, y, c='k', s=5, edgecolor='')
    
    ax.set_title("Kernel Density Function - Mountainside, NJ")
    ax.set_xlabel("Normalized Latitude")
    ax.set_ylabel("Normalized Longitude")
    
    return kernel
    
print()
    
fig, axarr = plt.subplots(1, 2)
ax = axarr[0]
k1 = kde1(gn['x']-np.average(gn['x']),gn['y']-np.average(gn['y']),ax)
fig.set_size_inches(25,25)
fig.suptitle('Kernel Density Function - Mountainside, NJ')
plt.xlabel('Normalized Latitude')
plt.ylabel('Normalized Longitude')
# https://gist.github.com/daleroberts/7a13afed55f3e2388865b0ec94cd80d2

plt.show()




```

```python
redraw = k1.resample(333)
```
